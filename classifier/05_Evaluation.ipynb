```python
import numpy as np
import matplotlib.pyplot as plt

from code.util import *
```

```python
ml = load("data/cv_scores.pkl")
ml.keys()
```

```python
m1 = np.load("cv/m1_cv.npy")
m2 = np.load("cv/m2_cv.npy")
m3 = np.load("cv/m3_cv.npy")
f123 = np.load("cv/fusion_123_cv.npy")
```

```python
acc = [
    ml["MultinomialNB"]["test_accuracy"],
    ml["LGBMClassifier"]["test_accuracy"],
    ml["LogisticRegression"]["test_accuracy"],
    list(m1[0]),
    list(m2[0]),
    list(m3[0]),
    list(f123[0])
]
```

```python
rec = [
    ml["MultinomialNB"]["test_recall"],
    ml["LGBMClassifier"]["test_recall"],
    ml["LogisticRegression"]["test_recall"],
    list(m1[2]),
    list(m2[2]),
    list(m3[2]),
    list(f123[2])
]
```

```python
labels = [
    "Naive Bayes",
    "LGBM",
    "Logistic Regression",
    "MLP",
    "CNN",
    "LSTM",
    "Fusion Network"
]
```

```python
plt.figure(figsize=(12, 10))
plt.boxplot(acc)
plt.xticks(ticks=range(1, 8), labels=labels, rotation=90)
plt.title("Accuracy by Model")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.boxplot(rec)
plt.xticks(ticks=range(1, 8), labels=labels, rotation=90)
plt.title("Recall by Model")
plt.ylabel("Recall")
plt.show()
```
